```python
from tabula import read_pdf
import pandas as pd
```

```python
path = 'maharastra.pdf'
```

```python

df = read_pdf(path)[0]
```

```python
df
```

```python
all_null_indexes = (df.isna().T.min().astype(int) * df.index).values
# remove_zeros
all_null_indexes = all_null_indexes[all_null_indexes != 0]
all_null_indexes
```

```python
df_flt = df[all_null_indexes[0]+2:]
df_flt
```

```python
# Create Datastore
clean_store = {c:[] for c in df.columns}
clean_store
```

```python
temp_store = {c:'' for c in df.columns}
temp_store
```

```python
for idx, row in df_flt.iterrows():    
    for column in temp_store:
        if not pd.isna(row[column]):
            if temp_store[column] != '':
                if temp_store[column][-1] != '$':
                    temp_store[column] += '$'
            if not 'É' in row[column]:
                temp_store[column] += row[column]
       
    # Condition when data checked in
    latlon = temp_store['Unnamed: 3']
    if ('N' in latlon and 'E' in latlon) or '--' in latlon:
        # Temp store to clean store
        for column in temp_store:
            clean_store[column].append(temp_store[column])
        # Cleanup for next loop
        temp_store = {c:'' for c in df.columns}
```

```python
clean_df = pd.DataFrame(clean_store)
clean_df
```

```python
# Clean Name
## Remove Trailing $
clean_df['Unnamed: 0'] = clean_df['Unnamed: 0'].str.rstrip('$')
## Replace $ with ,
clean_df['Unnamed: 0'] = clean_df['Unnamed: 0'].str.replace('$', ',')
```

```python
# Clean Latlong
## Remove Trailing $
clean_df['Unnamed: 3'] = clean_df['Unnamed: 3'].str.rstrip('$')
## Replace $ with <space>
clean_df['Unnamed: 3'] = clean_df['Unnamed: 3'].str.replace('$', ' ')
```
